```python
import numpy as np
import torch
from sklearn.manifold import TSNE

import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.cm as cm
import pickle
from collections import defaultdict
```

```python
class image_props():
    def __init__(self,tensor,qtype):
        self.feats = tensor
        self.qtypes = [qtype]
    
    def _add_qtype(self,qtype):
        self.qtypes.append(qtype)
```

```python
# qtype = "YesNo"
qtypes_list = ["choose", "color", "number", 'shape', "Others", "YesNo"]

qtypes_dict = {}
for i,qtype in enumerate(qtypes_list):
    file = "../../../../data/saved_embed2/{}_image_ids.pkl".format(qtype)
    qtype_index = pickle.load(open(file,'rb'))
    file = "../../../../data/saved_embed2/{}.pt".format(qtype)
    qtype_tensor = torch.load(file)
    print(qtype, qtype_tensor.shape)
    for index,tensor in zip(qtype_index,qtype_tensor):
        if index.item() not in qtypes_dict:
            qtypes_dict[index.item()] = image_props(tensor,qtype)
        else:
            qtypes_dict[index.item()]._add_qtype(qtype)
#     qtypes_dict[qtype] = (i,qtype_tensor)
    
#     for idx in qtype_index:
# #         print(idx.item())
#         qtypes_dict[idx.item()] += 1
    
# #     qtypes_dict[qtype] = (i,qtype_tensor)
```

```python
color_each_class = cm.rainbow(np.linspace(0,1,6))
qtypes_list = ["choose", "color", "number", "shape", "Others", "YesNo"]

qtypes_dict = {}
for i,qtype in enumerate(qtypes_list):
    file = "../../../../data/saved_embed/{}.pt".format(qtype)
#     qtype_feat = pickle.load(open(file,'rb'))
#     file = "../../../../data/saved_embed/{}.pt".format(qtype)
    qtype_tensor = torch.load(file)
    qtypes_dict[qtype] = (i, qtype_tensor)
#     print(qtype, qtype_tensor.shape)
#     for index,tensor in zip(qtype_index,qtype_tensor):
#         if index.item() not in qtypes_dict:
#             qtypes_dict[index.item()] = image_props(tensor,qtype)
#         else:
#             qtypes_dict[index.item()]._add_qtype(qtype)

r_idx = torch.randint(high=600,size=(500,))

qtype_feats = None
qtype_labels = None
for qtype in qtypes_dict.keys():
    if qtype_feats is None:
        i,qtype_feats = qtypes_dict[qtype]
        qtype_feats = qtype_feats[r_idx]
        qtype_labels = torch.tensor(color_each_class[i]).unsqueeze(0).repeat(qtype_feats.shape[0],1)
    else:
        i,qtype_feat = qtypes_dict[qtype]
        qtype_feat = qtype_feat[r_idx]
        qtype_feats = torch.cat((qtype_feats,qtype_feat))
        print(color_each_class[i])
        qtype_labels = torch.cat((qtype_labels,torch.tensor(color_each_class[i]).unsqueeze(0).repeat(qtype_feat.shape[0],1)))
print(qtype_feats.shape)
print(qtype_labels.shape)
```

```python
tsne = TSNE(n_components=2,n_iter=12000,n_jobs=-1,init='pca',perplexity=50)



qtype_features = np.asarray(qtype_feats, dtype='float64')

qtype_tsne = tsne.fit_transform(qtype_features)
```

```python
# fig = plt.figure()
# ax = fig.add_subplot(projection='3d')
color_each_class = cm.rainbow(np.linspace(0,1,6))
cell_text = None
for i,qtype in enumerate(qtypes_dict.keys()):
    if cell_text is None:
        cell_text = np.expand_dims(["",qtypes_list[qtypes_dict[qtype][0]]],axis=0)
    else:
        cell_text = np.concatenate((cell_text,np.expand_dims(["",qtypes_list[qtypes_dict[qtype][0]]],axis=0)),axis=0)
legend = plt.table(cellText=cell_text,colWidths=[0.05,0.12],cellLoc='right',loc='best')
for i,qtype in enumerate(qtypes_dict.keys()):
    legend[(i,0)].set_facecolor(color_each_class[qtypes_dict[qtype][0]])
plt.rcParams["figure.figsize"] = (15,10)

plt.scatter(qtype_tsne[:,0],qtype_tsne[:,1],c=qtype_labels)
```
